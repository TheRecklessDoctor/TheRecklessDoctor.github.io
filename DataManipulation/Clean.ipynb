```python
# import required libraries
import pandas as pd
```

```python
# import dataset
data = pd.read_csv("../data/world_population.csv")
data.head()
```

```python
data = data.replace("DR Congo","Democratic Republic of the Congo")
data
```

```python
GR_ds = data[["Country","Continent","Growth Rate"]]
GR_ds.head()
```

```python
sorted_GR = GR_ds.sort_values(by=["Growth Rate"],ascending=False)
sorted_GR.head()
```

```python
#export to csv file
sorted_GR.iloc[:11,:].to_csv("../data/Top10Growth_Rate.csv")
```

```python
#create new dataset for population changes over the years
copy1 = data.copy(deep=True)

copy1["change2022_1970"] = copy1["2022 Population"] - copy1["1970 Population"]

copy1.to_csv("../data/change.csv")
```

```python
# dataset for line chart

ds_copy = data.copy(deep=True)
grouped = ds_copy.groupby("Continent").mean()
grouped
```

```python
# modify the dataset for the line chart

grouped = grouped.reset_index() #reset the index from Continent to a numerical one
usefulcols = grouped[["Continent","2022 Population","2020 Population","2015 Population","2010 Population","2000 Population","1990 Population","1980 Population"]]
usefulcols.head()
```

```python
# melt the columns to get dates as a seperate column.
melted = usefulcols.melt(id_vars=["Continent"], var_name="Date",value_name="Value")
melted
```

```python
melted = melted.replace(["2022 Population","2020 Population","2015 Population","2010 Population","2000 Population","1990 Population","1980 Population",],["2022","2020","2015","2010","2000","1990","1980"])
melted
```

```python
# convert to datetime
melted['Date'] = pd.to_datetime(melted["Date"],format = "%Y")
melted
```

```python
melted.to_csv("../data/lineChart.csv")
```
